# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`. You should've created these data files in the previous section: *Fixing Data Types Pt 3*.

```python
# load datasets
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

```python
df_08.hist(figsize = (8,8))
plt.show()
```

```python
df_18.hist(figsize = (8,8))
plt.show()
```

```python
plt.figure(figsize = (11,7))
df_08_corr = df_08.corr()
sns.heatmap(df_08_corr, annot = True)
plt.title('Correlation heatmap between attributes in 2008')
plt.show()
```

```python
df_18_corr = df_18.corr()
df_18_corr
```

```python
pd.plotting.scatter_matrix(df_18, figsize = (18,15))
plt.show()
```
